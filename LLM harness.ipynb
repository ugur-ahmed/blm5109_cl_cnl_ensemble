```python
import json
from sklearn.model_selection import train_test_split

# Dosyayı yükle
with open('/content/arc_eval.json', 'r', encoding='utf-8') as f:
    data = [json.loads(line) for line in f]

# Veriyi %80 eğitim, %20 test olarak ayır
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)

# Veriyi "train" ve "test" bölümleri ile tek bir JSON dosyasında kaydet
dataset = {
    "train": train_data,
    "test": test_data
}

# Tek JSON dosyasını oluştur
with open('/content/arc_eval_test.json', 'w', encoding='utf-8') as f:
    json.dump(dataset, f, indent=2, ensure_ascii=False)
```
